```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

## 1. Tutorial on Iris Dataset

```python
# 1. Load the dataset
iris = load_iris()
X = iris.data
y = iris.target

# 2. Split the data into training and testing sets
# 70% training, 30% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Create and train the KNN model with k=5
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# 4. Make predictions
y_pred = knn.predict(X_test)

# 5. Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Optional: Classification report for precision, recall, f1-score
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

## 2. Apply KNN algorithm on the wine dataset and observe accuracy on training and test dataset with different values of ‘K’(2 to 5)

```python
# Import required libraries
from sklearn.datasets import load_wine


# 1. Load the Wine dataset
wine = load_wine()
X = wine.data
y = wine.target

# 2. Split into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Train KNN for K=2 to 5 and evaluate accuracy
print("Wine Dataset - KNN Accuracy:")
for k in range(2, 6):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)

    # Training accuracy
    y_train_pred = knn.predict(X_train)
    train_acc = accuracy_score(y_train, y_train_pred)

    # Test accuracy
    y_test_pred = knn.predict(X_test)
    test_acc = accuracy_score(y_test, y_test_pred)

    print(f"K={k} --> Training Accuracy: {train_acc:.2f}, Test Accuracy: {test_acc:.2f}")
```

## 6. Perform all above que(2 to 5) using 10 fold cross validation and observe accuracy on training and test datasetwith different values of ‘K’(2 to 5)

```python
import warnings
warnings.filterwarnings('ignore')  # suppress warnings

import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris, load_wine, load_breast_cancer






# Example with multiple datasets
datasets = {
    'Iris': load_iris(),
    'Wine': load_wine(),
    'Breast Cancer': load_breast_cancer(),

}



for name, dataset in datasets.items():
    X = dataset.data
    y = dataset.target

    print(f"\n{name} Dataset - 10-Fold Cross-Validation Accuracy:")
    for k in range(2, 6):
        knn = KNeighborsClassifier(n_neighbors=k)

        # 10-fold cross-validation
        scores = cross_val_score(knn, X, y, cv=10)  # default scoring='accuracy'
        mean_acc = scores.mean()

        print(f"K={k}: CV Accuracy={mean_acc:.2f}")
```

```python
from sklearn.model_selection import cross_val_score, KFold

file_path = '/content/drive/MyDrive/Datsets/cleaned_dataset_Thyroid1.csv'
df= pd.read_csv(file_path)
# Ensure numeric values
df = df.apply(pd.to_numeric, errors='coerce').dropna()

# Features and target
X = df.drop('binaryClass', axis=1).values
y = df['binaryClass'].values.astype(int)

# 10-fold cross-validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)

print("Thyroid Dataset - 10-Fold Cross-Validation Accuracy:")
for k in range(2, 6):
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, y, cv=kf)
    print(f"K={k}: CV Accuracy={scores.mean():.2f}")
```

```python
# Load Seed dataset (assuming it's already a CSV or TXT loaded into pandas)
file_path = '/content/drive/MyDrive/Datsets/seeds_dataset.txt'
seeds = pd.read_csv(file_path, sep='\s+', header=None)
seeds.columns = ['area','perimeter','compactness','length_of_kernel','width_of_kernel',
                 'asymmetry_coefficient','length_of_groove','class']

# Separate features and target
X = seeds.drop('class', axis=1).values
y = seeds['class'].values

# 10-fold cross-validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)

print("Seed Dataset - 10-Fold Cross-Validation Accuracy:")
for k in range(2, 6):
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, y, cv=kf)
    print(f"K={k}: CV Accuracy={scores.mean():.2f}")
```
